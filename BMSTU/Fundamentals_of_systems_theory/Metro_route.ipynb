```python
import pandas as pd
import numpy as np
import math
```

```python
# Функция для нахождения вершины с минимальным расстоянием
def min_distance(dist, visited):
    min_dist = np.inf
    min_index = -1
    for v in range(len(dist)):
        if dist[v] < min_dist and not visited[v]:
            min_dist = dist[v]
            min_index = v
    return min_index

# Функция для нахождения кратчайшего пути
def dijkstra_algorithm(graph, start, end):
    # Инициализация переменных
    dist = [np.inf] * len(graph)
    dist[start] = 0
    visited = [False] * len(graph)
    prev = [-1] * len(graph)

    # Выполнение алгоритма
    for i in range(len(graph)):
        v = min_distance(dist, visited)
        visited[v] = True
        if v == end:
            break
        for j in range(len(graph)):
            if graph[v][j] != 0 and not visited[j]:
                if dist[v] + graph[v][j] < dist[j]:
                    dist[j] = dist[v] + graph[v][j]
                    prev[j] = v

    # Составление пути
    path = [end]
    while prev[path[-1]] != -1:
        path.append(prev[path[-1]])
    path.reverse()

    # Возвращение результата
    return dist[end], path
```

```python
df = pd.ExcelFile("Data_Metro.xlsx")
df = df.parse()
df = df.drop(columns='Unnamed: 0')
stations = list(df.columns)
df.head()
```

```python
i = 0
for column in df:
    df[str(column)][i] = 0
    i += 1
df.head()
```

```python
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        df.iloc[i,j] = df.iloc[j,i]
df.head()
```

```python
df.replace(np.nan, math.inf, inplace=True)
df.head()
```

```python
df_metro = df

print("Введите маршрут")
print("От:")
start_station = input()
print("До:")
end_station = input()

distance, path = dijkstra_algorithm(df_metro.values, stations.index(start_station), stations.index(end_station))
print()
print("Кратчайшее время в пути:", distance, "минут")
print("Кратчайший путь:")
for el in path:
    print(stations[el], end=' -> ')
```
